#MACHINE LEARNING

```python
#!pip install scikit-learn

from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rc = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)
sfs = SequentialFeatureSelector(rc, n_features_to_select=30, direction='forward', cv=split)
```

```python
remove_columns = ['season', 'data', 'ganhador', 'target', 'time', 'time_opp']
```

```python
selected_columns = df.columns[~df.columns.isin(remove_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
df
```

```python
sfs.fit(df[selected_columns], df['target'])
```

```python
predictors = list(selected_columns[sfs.get_support()])
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1):
  predicoes = []

  seasons = sorted(data['season'].unique())

  for i in range(start, len(seasons), step):
    season = seasons[i]

    train = data[data['season'] < season]
    test = data[data['season'] == season]

    model.fit(train[predictors], train['target'])

    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index)

    combined = pd.concat([test['target'], preds], axis=1)
    combined.columns = ['final','predicao']

    predicoes.append(combined)
  return pd.concat(predicoes)
```

```python
sorted(df['season'].unique())
```

```python
predicoes = backtest(df, rc, predictors)
```

```python
predicoes
```

```python
from sklearn.metrics import accuracy_score

predicoes = predicoes[predicoes['final'] != 2]
accuracy_score(predicoes['final'], predicoes['predicao'])
```

```python
def find_team_average(team):
  rolling = team.rolling(10).mean()
  return rolling

df_rolling = df_rolling.groupby(['time','season'], group_keys=False).apply(find_team_average)
```

```python
df_rolling
```

```python
rolling_cols = [f'{col}_10' for col in df_rolling.columns]
df_rolling.columns = rolling_cols

df = pd.concat([df, df_rolling], axis=1)
```

```python
df
```

```python
df = df.dropna()
df
```

```python
def shift_col(time, col_name):
  next_col = time[col_name].shift(-1)
  return next_col

def add_col(df, col_name):
  return df.groupby('time', group_keys=False).apply(lambda x: shift_col(x, col_name))

df['home_next'] = add_col(df, 'home')
df['time_opp_next'] = add_col(df, 'time_opp')
df['data_next'] = add_col(df, 'data')
```

```python
df = df.copy()
```

```python
full = df.merge(
    df[rolling_cols + ['time_opp_next','data_next','time']],
    left_on = ['time','data_next'],
    right_on = ['time_opp_next','data_next']
)
```
